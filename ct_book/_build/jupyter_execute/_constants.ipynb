# Download the datasets to create the hash

The SHA256 checksums are currently not published, create them as well.

```python
import requests
import hashlib
import os
```

```python
def download_file(url, file_path):
    """
    Downloads a file from a URL to the specified file path.

    Parameters:
        url (str): The URL of the file to download.
        file_path (str): The local path where the file will be saved.

    Returns:
        None
    """
    os.makedirs(os.path.dirname(file_path), exist_ok=True)
    if not os.path.exists(file_path):
        print(f"Downloading: {file_path}")
        response = requests.get(url, stream=True)
        if response.status_code == 404: # No file
            return None
        response.raise_for_status()  # Raise HTTPError for bad responses
        
        with open(file_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
        print(f"File downloaded successfully: {file_path}")
    sha256 = hashlib.sha256()
    with open(file_path, 'rb') as file:
        while chunk := file.read(8192):
            sha256.update(chunk)
    sha = sha256.hexdigest()
    print(f"{file_path}: {sha}")
    return sha

gases = ["ch4", "n2o", "co2", "co2e_100yr"]
sectors = ['agriculture',
 'buildings',
 'fluorinated-gases',
 'forestry-and-land-use',
 'fossil-fuel-operations',
 'manufacturing',
 'mineral-extraction',
 'power',
 'transportation',
 'waste']

res = {
    
}

for gas in gases:
    if gas not in res:
        res[gas] = {}
    for sector in sectors:
        sector2 = sector.replace("-", "_")
        url_template = "https://downloads.climatetrace.org/v3/sector_packages/{gas}/{s}.zip"
        file_path = f"/tmp/ct_raw/{gas}/{sector}.zip"
        sha = download_file(url_template.format(gas=gas,s=sector), file_path)
        if sha is None:
            sha = download_file(url_template.format(gas=gas,s=sector2), file_path)
        res[gas][f"{sector}.zip"] = f"sha256:{sha}" if sha is not None else None

print(res)
```
